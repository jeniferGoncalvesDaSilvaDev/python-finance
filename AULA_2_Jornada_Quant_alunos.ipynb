Verificação de empresas que ainda são negociadas

```python
from datetime import datetime, timedelta

# Function to check if a stock is delisted
def is_delisted(symbol):
    try:
        # Get historical data for the stock
        stock = yf.Ticker(symbol)
        historical_data = stock.history(period='1mo')

        # Get the last trading day
        last_trading_day = historical_data.index[-1].date()

        # Check if the last trading day is not yesterday
        yesterday = datetime.now()
        return last_trading_day != yesterday.date()
    except Exception:
        # If there is an exception (e.g., stock not found), consider it as delisted
        return True
```

```python
divida_bruta_patr['ACAO'] = divida_bruta_patr.index + '.SA'
divida_bruta_patr['Delisted'] = divida_bruta_patr['ACAO'].apply(is_delisted).head(50)
```

```python
empresas_filtradas = divida_bruta_patr[divida_bruta_patr['Delisted'] == False]
```

```python
empresas_filtradas = empresas_filtradas.sort_values('roe', ascending = False).head(10)
empresas_filtradas
```

```python
plt.figure(figsize=(10,6));
plt.bar(empresas_filtradas.tail(10).index, empresas_filtradas.tail(10).roe);
plt.xticks(rotation=45);
plt.title("Ações selecionadas por Divida Bruta/Pat Liquido e ROE");
```

## Modelos de Otimização

O que são modelos de otimização de carteira?

Os modelos de otimização de carteira são ferramentas analíticas que ajudam os investidores a tomar decisões sobre como alocar seus recursos financeiros entre diferentes ativos, como ações, títulos, fundos mútuos, entre outros, de maneira a maximizar o retorno esperado para um nível de risco específico ou minimizar o risco para um determinado nível de retorno.

O objetivo fundamental desses modelos é criar uma carteira que equilibre os diferentes ativos de forma a atingir os objetivos do investidor, considerando suas preferências de risco e retorno.

É importante notar que, embora esses modelos forneçam valiosas insights, eles estão sujeitos a algumas limitações, como a premissa de que os retornos passados são indicativos de retornos futuros, a sensibilidade às estimativas de parâmetros e a suposição de distribuição normal dos retornos. Investidores devem estar cientes dessas limitações ao utilizar modelos de otimização de carteira e considerar informações adicionais ao tomar decisões de investimento.

### Importando bibliotecas

```python
import riskfolio as rp
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
#Parâmetros

#Data
inicio='2023-01-01'
fim='2023-09-30'

#Seleção dos ativos

ativos = ['CORR3.SA', 'ODPV3.SA', 'CXSE3.SA', 'BRAP3.SA', 'CEBR3.SA', 'PSSA3.SA','ITUB3.SA', 'PINE3.SA', 'BSLI3.SA', 'LIPR3.SA']
```

```python
#Extrair os dados de preço

carteira = yf.download(ativos, start=inicio, end=fim)['Adj Close']
```

```python
carteira.head()
```

```python
#Calcular os retornos dos ativos

retornos = carteira.pct_change().dropna()
retornos
```

```python
retornos.cov()
```

##Modelo de otimização - Markowitz MAX SHARPE

Sharpe Ratio = (Retorno Esperado - Taxa Livre de Risco)/Volatilidade da Carteira

```python
portfolio = rp.Portfolio(returns=retornos)

mu = 'hist' #estimando os retornos com base no histórico
cov = 'hist' #estimar a matriz de covariância através do histórico

portfolio.assets_stats(method_mu=mu, method_cov=cov, d=0.94)

#Construindo o modelo

model='Classic'
rm='MV'
obj='Sharpe'
hist=True
rf=0
l=0

w = portfolio.optimization(model=model, rm=rm, obj=obj, rf=rf, l=l, hist=hist)

display(w.T)
```

```python
ax = rp.plot_pie(w=w,
                 title= 'Otimização de Portfólio',
                 others=0.01,
                 nrow=25,
                 cmap='tab20',
                 height=8,
                 width=10,
                 ax=None)
```
